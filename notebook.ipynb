```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def mon_inventaire(inventaire):
    print("--- État du Stock --")
    for article, quantite in inventaire.items():
        print(f"{article.capitalize()} : {quantite}")
    print("---------------------")

# Test de la fonction
mon_inventaire(stock)
```

```python
def fabriquer(produit, quantite):
    if produit in stock:
       #on ajoute par contre si il ne l'est pas
        stock[produit] += quantit
        print(f"Succès : {quantite} {produit} ajouté(s) au stock.")
    else:
        print(f"Erreur : Le produit '{produit}' n'existe pas dans l'inventaire.")

# on teste notre fonction
produit_choisi = input("Que voulez-vous fabriquer ? ")
nombre = int(input("En quelle quantité ? "))

fabriquer(produit_choisi, nombre)

# On vérifie le résultat avec la fonction de visualisation précédente
afficher_inventaire(stock)
```
